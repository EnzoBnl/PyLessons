```python
#import
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import scipy.interpolate as interp
from random import randint
#2D plot :
#add a courb in other figure 
figure2, axes_figure2 = plt.subplots()
abscisses = list(range(-10,11))
ordinates = [x**2 for x in abscisses]
#axes_figure2.plot(abscisses,ordinates)
axes_figure2.hist2d([randint(-100,100)*i+(300000 if i%2==0 else 0) for i in range(10000)],[randint(-100,100)*i for i in range(10000)],100)
#create a figure :
figure = plt.figure()
#create axes on this figure
axes = figure.add_subplot(111, projection='polar')
#create abscisses and ordinates
abscisses = list(range(-10,11))
ordinates = [x**2 for x in abscisses]
#plot them in our axes
axes.plot(abscisses,ordinates)

#plt.subplots()
figure = plt.figure()
#add a courb in one of the axes :
#create a list of abscisses 
abscisses = list(range(-10,11))
#create a list of ordinates 
ordinates = [x**2 for x in abscisses]
axes1=figure.add_subplot(2,2,1)
axes1.plot(abscisses,ordinates)
#add another courb on the same axes
#create a list of abscisses using np.arange
abscisses = np.arange(-5,5,0.1)
#create a list of ordinates 
ordinates = [x**3 for x in abscisses]
axes1.plot(abscisses,ordinates)
#add a courb in other axes 
abscisses = np.arange(-5,5,0.1)
ordinates = [3*x+5 for x in abscisses]
axes4 = figure.add_subplot(2,2,4)
axes4.plot(abscisses,ordinates)
#add a 3D plot on axes of the first figure
axes3 = figure.add_subplot(2, 2, 3, projection='3d')
from random import randint as r
axes3.plot([r(0,100) for _ in range(20)],[r(0,100) for _ in range(20)],[r(0,100) for _ in range(20)],'o')
#add histogram
axes2 = figure.add_subplot(2,2,2)
axes2.hist([r(0,10)*i**2 for i in range(200)], 50)
axes2.hist([r(0,2000)*i for i in range(200)], 50,alpha=0.5)
plt.show()
#https://matplotlib.org/api/markers_api.html
figure.savefig("quadriplot.png")
```
